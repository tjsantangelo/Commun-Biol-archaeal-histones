### After importing PyRosetta, it is important to clean up the pdb structure.

```python
from pyrosetta.toolbox import cleanATOM
cleanATOM("pdb/5T5K.pdb")
```

```python
cpose = pose_from_pdb("pdb/5T5K.clean.pdb")
```

```python
print(cpose.pdb_info())
```

```python
from pyrosetta.rosetta.protocols.loops import get_fa_scorefxn
sfxn = get_fa_scorefxn()
```

```python
sfxn.show(cpose)
```

```python
temp_pose = cpose.clone()
sfxn.show(temp_pose)
```

### We need to relax the structure to optimize the conformations for point mutations.

```python
movemap = MoveMap()
movemap.set_bb(False)
movemap.set_chi(True)
relax = pyrosetta.rosetta.protocols.relax.FastRelax()
relax.constrain_relax_to_start_coords(True)
relax.coord_constrain_sidechains(True)
relax.ramp_down_constraints(False)
relax.set_scorefxn(sfxn)
relax.set_movemap(movemap)
relax.apply(temp_pose)
temp_pose.dump_pdb('pdb/5T5K.relax.pdb')
```

```python
rpose = pose_from_pdb('pdb/5T5K.relax.pdb')
sfxn.show(rpose)
```

### Observation of the scorefunction before and after relaxing the structure shows a significant reduction in the overall free energy, important for finding optimal conformations for point mutations.

```python
from pyrosetta.toolbox import mutate_residue
```

```python
R20S = pyrosetta.pose_from_pdb("pdb/5T5K.relax.pdb")
mutate_residue(R20S, 19, "S")
sfxn.show(R20S)
```

```python
T55L = pyrosetta.pose_from_pdb("pdb/5T5K.relax.pdb")
mutate_residue(T55L, 54, "L")
sfxn.show(T55L)
T55L.dump_pdb("pdb/5T5K.T55L.pdb")
```

### Because the free energy associated with the T55L variant is significantly greater than WT (rpose), we need to relax the structure to eliminate any incompatible conformations.

```python
temp_T55L = pyrosetta.pose_from_pdb("pdb/5T5K.T55L.pdb")
movemap = MoveMap()
movemap.set_bb(False)
movemap.set_chi(True)
relax = pyrosetta.rosetta.protocols.relax.FastRelax()
relax.constrain_relax_to_start_coords(True)
relax.coord_constrain_sidechains(True)
relax.ramp_down_constraints(False)
relax.set_scorefxn(sfxn)
relax.set_movemap(movemap)
relax.apply(temp_T55L)
temp_T55L.dump_pdb("pdb/5T5K.relax_T55L.pdb")
```

```python
rT55L = pyrosetta.pose_from_pdb("pdb/5T5K.relax_T55L.pdb")
sfxn.show(rT55L)
```

```python
G17D = pose_from_pdb('pdb/5T5K.relax.pdb')
mutate_residue(G17D, 16, "D")
sfxn.show(G17D)
```

```python
G17D.dump_pdb("pdb/5T5K.G17D.pdb")
```

```python
rG17D = pyrosetta.pose_from_pdb("pdb/5T5K.relax_G17D.pdb")
sfxn.show(rG17D)
```
